```python
import pandas as pd

# ini load dataset
df= pd.read_csv("food_coded.csv")

# ini baca 5 paling atas, buat mastiin datanya jalan
df.head()
```

```python
# cek info dataset
df.info()

# saya cek ada beberapa yang kosong
df.isnull().sum()
```

```python
# Disini saya makesure lagi secara keseluruhan total missing values
missing_cols = df.columns[df.isnull().any()]
print(df[missing_cols].isnull().sum())
```

```python
# disini saya cek missing value dalam bentuk persen
missing_percent = (df[missing_cols].isnull().sum() / len(df)) * 100
print(missing_percent)
```

```python
# Imputasi data numerik dengan mean
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())

# Imputasi data kategorikal dengan modus
cat_cols = df.select_dtypes(include=['object']).columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

# memastikan sudah selsai masalahnya
print(df.isnull().sum().sum())  
```

```python
df.info()
```

```python
# Ubah kolom numerik yang masih object ke float, yaitu GPA
df['GPA'] = pd.to_numeric(df['GPA'], errors='coerce')
df['weight'] = pd.to_numeric(df['weight'], errors='coerce')

# MENGECEK hasilnya kembali
df[['GPA', 'weight']].info()

```

```python
# export ke csv dimana datanya udh clean
df.to_csv('food_coded_clean.csv', index=False)
```
